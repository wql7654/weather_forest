K-NEAREST NEIGHBORS (KNN)

KNN : K 개의 가장 가까운 레이블이 붙은 데이터 포인트를 확인하십시오.
Classification method.
먼저  train 데이터가 필요합니다. train=fit
fit () : 데이터를 맞추고 데이터를 조정합니다.
predict () : 데이터를 예측합니다.
KNN이 무엇인지 이해하지 못한다면 https://www.youtube.com/watch?v=CyuI2F_wJWw
x : features
y :  target variables(normal(정상), abnormal(비정상))
n_neighbors : K.이 예에서는 3입니다. 가장 가까운 3 개의 레이블이 붙은 데이터 요소를 봅니다.

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
knn.fit(x,y)
prediction = knn.predict(x)
print('Prediction: {}'.format(prediction))
```

데이터를 맞추고 KNN으로 예측합니다.
따라서 우리는 정확성을 예측하거나 정확도가 정확한지 아니면 정확성이 우리의 결과를 평가하는 데 가장 좋은 척도인지에 대해 답할 수있게 해줍니다.
모델 성능 측정 :
정확한 예측의 일부인 정확도가 일반적으로 메트릭으로 사용됩니다. 우리는 그것을 사용할 것입니다. 그러나 또 다른 문제가 있습니다.

보시다시피 저는 x (피처)로 데이터를 학습하고 x (피처)를 다시 예측합니다. 예, 당신은 바르게 읽고 있습니다. 그렇습니다. 다시 말하자면, 그것은 어리석은 짓입니다. :)


따라서 데이터 열과 테스트 세트를 분리해야합니다.

train :fitting set train사용
test :  test set에 prediction을 만듬
train와 est sets의 경우  fitted data 와  tested data는 전혀다릅니다.
train_test_split (x, y, test_size = 0.3, random_state = 1)
x : features
y :  target variables (normal,abnormal)(정상, 비정상)
test_size : 테스트 크기의 백분율. 예제 test_size = 0.3, 테스트 크기 = 30 % 및 train 크기 = 70 %
random_state : 시드를 설정합니다. 이 시드가 동일한 숫자 인 경우 train_test_split ()은 매번 똑같은 분할을 생성합니다.
fit (x_train, y_train) : train sets에 fit합니다.
score (x_test, y_test)) : 테스트 세트에 대한 예측 및 정확성 제공

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
#print('Prediction: {}'.format(prediction))
print('With KNN (K=3) accuracy is: ',knn.score(x_test,y_test)) # accuracy
```

정확도가 86 %이므로 좋습니다. 이게 좋은지는 마직막 결과를 보고 평가를 하겠습니다.
이제 질문은 K = 3을 선택하거나 K를 선택하는 데 필요한 값을 선택하는 것입니다. 답은 모델의 복잡성입니다


모델 복잡성 :

K는 일반 이름입니다. 그것은 hyperparameter라고 부릅니다. 지금은 단지 K가 하이퍼 파라미터라는 것을 알기 때문에
최적의 성능을 제공하는 하이퍼 파라미터를 선택해야합니다.
자료에 따르면 k가 작으면 모형이 복잡한 모형이 과잉으로 이어질 수 있다고합니다. 
이는 모델이 train set을 암기하고 정확한 정확도로 시험 세트를 예측할 수 없다는 것을 의미합니다.
k가 크다면 덜 복잡한 모델 인 모델은 언더 피트로 이어질 수 있습니다.
아래에서 1부터 25까지의 K 값 범위 (제외)와 각 K 값의 정확도를 찾습니다. 
플롯에서 볼 수 있듯이 K가 1 일 때 train se를 기억하고 테스트 세트 (overfit)에 대해 정확한 정확성을 부여 할 수 없습니다.
또한 K가 18이라면 모델은 언더 피트로 이어집니다. 다시 한 번 정확성만으로는 충분하지 않습니다. 
그러나 K가 18 일 때 (최상의 성능), 정확도는 거의 88 %에 이릅니다.

```python
# Model complexity
neig = np.arange(1, 25)
train_accuracy = []
test_accuracy = []
# k 값이 다를경우 loop테스트
for i, k in enumerate(neig):
    # k (1~25)
    knn = KNeighborsClassifier(n_neighbors=k)
    # Fit with knn
    knn.fit(x_train,y_train)
    #train accuracy
    train_accuracy.append(knn.score(x_train, y_train))
    # test accuracy
    test_accuracy.append(knn.score(x_test, y_test))

#visualization
plt.figure(figsize=[13,8])
plt.plot(neig, test_accuracy, label = 'test 정확도')
plt.plot(neig, train_accuracy, label = 'tarin 정확도')
plt.legend()
plt.title('측정값 VS 정확도')
plt.xlabel('인접한 수')
plt.ylabel('정확도')
plt.xticks(neig)
plt.savefig('graph1.png')
plt.show()
print("최고의 정확도 is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```

이 정형 외과 환자 데이터는 회귀에 적절하지 않기 때문에 
pelvic_incidence의 비정상성과 sacral_sloper을 사용합니다.
 feature는 pelvic_incidence이고 target이 sacral_slope임
산점도를 보면 이해할 수있습니다.
reshape (-1,1) : x 또는 y 모양 (210,)을 사용하지 않고 sklearn에서 사용할 수 없으므로 
shape (-1,1)을 사용하고 x 또는 y의 모양은 (210, 1).

```python
# create data1 that includes pelvic_incidence that is feature and sacral_slope that is target variable
data1 = data[data['class'] =='Abnormal']
x = np.array(data1.loc[:,'pelvic_incidence']).reshape(-1,1)
y = np.array(data1.loc[:,'sacral_slope']).reshape(-1,1)
# Scatter
plt.figure(figsize=[10,10])
plt.scatter(x=x,y=y)
plt.xlabel('pelvic_incidence')
plt.ylabel('sacral_slope')
plt.show()
```

이제 회귀 분석을위한 데이터를 얻었습니다. 
회귀에서 문제는 목표 값이 price of house이나 sacral_slope와 같은 가변변수 이기때문에 지속적으로 변합니다.
이 점에 선을 맞춥니다.


선형 회귀

y = ax + b 여기서 y = target,, x =  feature 및 a = parameter of model
함수 (function)를 상실한 최소 오차 함수에 따라 모델 (a)의 파라미터를 선택한다.
선형 회귀 분석에서 우리는 손실 함수 (Least Square) (OLS)를 사용합니다.
OLS : 모든 잔차를 합산합니다. 그러나 일부 양수 및 음수 잔차는 서로 상쇄되므로 잔차의 제곱의 합을 구할 수 있습니다. OLS 라 불린다.
Score :Score는 (y_pred - y_mean) ^ 2) / (y_actual - y_mean) ^ 2 인 R ^ 2 방법을 사용합니다.

```python
# LinearRegression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
# Predict space
predict_space = np.linspace(min(x), max(x)).reshape(-1,1)
# Fit
reg.fit(x,y)
# Predict
predicted = reg.predict(predict_space)
# R^2 
print('R^2 score: ',reg.score(x, y))
# Plot regression line and scatter
plt.plot(predict_space, predicted, color='black', linewidth=3)
plt.scatter(x=x,y=y)
plt.xlabel('pelvic_incidence')
plt.ylabel('sacral_slope')
plt.show()
```

CROSS VALIDATION
KNN 방법에서 알다시피, 우리는 매번 정확하게 똑같이 나뉘는 random_state로 train test split을 사용합니다.
그러나 random_state를 사용하지 않으면 데이터가 매번 다르게 분할되고 분할 정확도에 따라 다릅니다.

예를 들어 데이터를 5 번 분할하고 맞추고 예측하는 경우 정확도는 각각 0.89, 0.9, 0.91, 0.92 및 0.93입니다. 
어떤 정확도를 사용합니까? 6 번 분할, 훈련 및 예측에서 정확도가 어느 정도인지 알 수 있습니까? 
대답은 잘 모르겠지만 교차 유효성 검사를 사용하면 허용되는 정확도를 찾을 수 있습니다.

교차 유효성 검사 (CV)

K folds = K fold CV.
K가 증가하면 계산량이 증가한다.
cross_val_score (reg, x, y, cv = 5) : 위에서 정의한 x와 y와 reg (선형 회귀)를 사용하고 K는 5입니다. 
이것은 5 번 (split, train, predict)

```python
# CV
from sklearn.model_selection import cross_val_score
reg = LinearRegression()
k = 5
cv_result = cross_val_score(reg,x,y,cv=k) # uses R^2 as score 
print('CV Scores: ',cv_result)
print('CV scores average: ',np.sum(cv_result)/k)
```

```python
정규 회귀 분석
선형 회귀 분석을 할 때 손실 함수를 최소화하면서 매개 변수 (계수)를 선택하십시오. 
선형 회귀 분석에서 피쳐 중 하나가 중요하다고 생각하면이 피쳐에 높은 계수를 부여합니다. 
그러나 이것은 KNN에 memorizing과 같은 과적합을 일으킬 수 있습니다. 
오버 피팅을 피하기 위해 큰 계수에 불이익을주는 정규화를 사용합니다.

리지 회귀 : 최초의 정규화 기법. 또한 L2 정규화라고도합니다.
릿지 회귀선 상실 fuction = OLS + alpha * sum (parameter^2)
알파는 우리가 적합하고 예측하기 위해 선택해야하는 매개 변수입니다. 
Picking alpha는 KNN에서 K picking와 비슷합니다. 
알다시피 알파는 최상의 정확도와 모델 복잡성을 위해 선택해야하는 하이퍼 매개 변수입니다. 
이 프로세스를 하이퍼 매개 변수 튜닝이라고합니다.

알파가 0이면? 잃어버린 함수 = OLS 그래서 선형 rigression입니다 :)
알파가 작아서 오버 피팅을 일으킬 수있는 경우
알파가 크면 언더 피팅이 발생할 수 있습니다. 그러나 작고 큰 것을 묻지 마십시오. 이들은 문제마다 변할 수 있습니다.
Lasso regression : 두 번째 정규화 기술. 또한 이것을 L1 정규화라고합니다.
Lasso regression lost fuction = OLS + alpha * sum (절대 값 (매개 변수))
데이터의 중요한 기능을 선택하는 데 사용할 수 있습니다. 
값이 0으로 축소되지 않는 지형지 물은 올가미 회귀에 의해 선택되기 때문에
기능을 선택하기 위해 회귀 데이터에 새로운 기능을 추가합니다.

https://ratsgo.github.io/machine%20learning/2017/05/22/RLR/
    
Linear vs Ridge vs Lasso First impression: Linear Feature Selection: 1.Lasso 2.Ridge Regression model: 1.Ridge 2.Lasso 
```

```python
# Ridge
from sklearn.linear_model import Ridge
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 2, test_size = 0.3)
ridge = Ridge(alpha = 0.1, normalize = True)
ridge.fit(x_train,y_train)
ridge_predict = ridge.predict(x_test)
print('Ridge score: ',ridge.score(x_test,y_test))
```

```python
# Lasso
from sklearn.linear_model import Lasso
x = np.array(data1.loc[:,['pelvic_incidence','pelvic_tilt numeric','lumbar_lordosis_angle','pelvic_radius']])
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 3, test_size = 0.3)
lasso = Lasso(alpha = 0.1, normalize = True)
lasso.fit(x_train,y_train)
ridge_predict = lasso.predict(x_test)
print('Lasso score: ',lasso.score(x_test,y_test))
print('Lasso coefficients: ',lasso.coef_)
```

이제는 정확성을 논의 할 수 있습니다. 모델 선택의 측정에 충분합니까?
예를 들어, 95 % 정상 및 5 % 비정상 샘플을 포함하는 데이터가 있고 우리 모델은 측정 기준에 대한 정확도를 사용합니다.
그러면 우리 모델은 모든 샘플에 대해 100 % 표준을 예측하고 정확도는 95 %이지만 모든 비정상 샘플은 잘못 분류합니다.
따라서 불균형 데이터에서 모델 측정 매트로 혼용 행렬을 사용해야합니다.
confusion matrix을 사용하는 동안 Random forest classifier를 사용하여 분류 방법을 다양화할 수 있습니다.
tp = true positive(20), fp = false positive(7), fn = false negative(8), tn = true negative(58)
tp = Prediction is positive(normal) and actual is positive(normal).
fp = Prediction is positive(normal) and actual is negative(abnormal).
fn = Prediction is negative(abnormal) and actual is positive(normal).
tn = Prediction is negative(abnormal) and actual is negative(abnormal)
precision = tp / (tp+fp)
recall = tp / (tp+fn)
f1 = 2 * precision * recall / ( precision + recall)

```python
# Confusion matrix with random forest
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
rf = RandomForestClassifier(random_state = 4)
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
print('Confusion matrix: \n',cm)
print('Classification report: \n',classification_report(y_test,y_pred))
```

```python
# visualization
sns.heatmap(cm,annot=True,fmt="d") 
plt.show()
```

 auc가 1에 가까워 지길 원합니다.

random forest classifier 및  5 fold cross validation
k 배 교차 검증에서, 데이터 세트는 k 개의 서브 세트들로 분할되고, 방법은 k 번 반복된다.
k 서브 세트 중 하나가 테스트 세트에 사용될 때마다 그리고 다른 것들은 트레이닝 세트이다. 
그 후 모든 k 시도에 대한 평균 정확도가 계산됩니다.

```python
random_state = np.random.RandomState(0)
clf = RandomForestClassifier(random_state=random_state) #RandomForestClassifier라이브러리
cv = StratifiedKFold(n_splits=5,shuffle=False) #StratifiedKFold 라이브러리
```

```python
clf
```

```python
cv
```

```python
# plot arrows
fig1 = plt.figure(figsize=[12,12])
ax1 = fig1.add_subplot(111,aspect = 'equal')
ax1.add_patch( # patch라이브러리
    patches.Arrow(0.45,0.5,-0.25,0.25,width=0.3,color='green',alpha = 0.5)
    )
ax1.add_patch(
    patches.Arrow(0.5,0.45,0.25,-0.25,width=0.3,color='red',alpha = 0.5)
    )

tprs = []
aucs = []
mean_fpr = np.linspace(0,1,100)
i = 1
for train,test in cv.split(x,y):
    prediction = clf.fit(x.iloc[train],y.iloc[train]).predict_proba(x.iloc[test])
    fpr, tpr, t = roc_curve(y[test], prediction[:, 1])
    tprs.append(interp(mean_fpr, fpr, tpr))
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    plt.plot(fpr, tpr, lw=2, alpha=0.3, label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))
    i= i+1

plt.plot([0,1],[0,1],linestyle = '--',lw = 2,color = 'black')
mean_tpr = np.mean(tprs, axis=0)
mean_auc = auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, color='blue',
         label=r'Mean ROC (AUC = %0.2f )' % (mean_auc),lw=2, alpha=1)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")
plt.text(0.32,0.7,'More accurate area',fontsize = 12)
plt.text(0.63,0.4,'Less accurate area',fontsize = 12)
plt.show()
```

```python
# ROC Curve with logistic regression
from sklearn.metrics import roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
# abnormal = 1 and normal = 0
data['class_binary'] = [1 if i == 'Abnormal' else 0 for i in data.loc[:,'class']]
x,y = data.loc[:,(data.columns != 'class') & (data.columns != 'class_binary')], data.loc[:,'class_binary']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
y_pred_prob = logreg.predict_proba(x_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.show()
```

```python
HYPERPARAMETER TUNING
KNN에서 언급했듯이 조정해야 할 하이퍼 파라미터가 있습니다.

예 :
KNN의 k
alpha at Ridge and Lasso
Random forest parameters  (예 : max_depth)
linear regression parameters (계수: coefficients)

Hyperparameter tuning:
다른 매개 변수의 모든 조합을 시도하십시오.
fit all of them
예측 성능 측정(measure prediction performance)
각각의 실적이 얼마나 좋은지 확인하십시오.
마지막으로 최상의 하이퍼 파라미터 선택
이 과정은이가장 어려운 부분입니다. 왜냐하면 모든 루프를 반복 작성하기 위해 많은 for 루프를 작성하기 때문입니다. 
GridSearchCV 인 한 라인 코드 만 있으면됩니다.
그리드 : K는 1에서 50까지입니다 (제외).
GridSearchCV는 knn과 grid를 가져 와서 그리드 검색을합니다.
모든 하이퍼 매개 변수의 조합을 의미합니다. 여기는 k입니다.
```

```python
# grid search cross validation with 1 hyperparameter
from sklearn.model_selection import GridSearchCV
grid = {'n_neighbors': np.arange(1,50)}
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, grid, cv=3) # GridSearchCV
knn_cv.fit(x,y)# Fit

# Print hyperparameter
print("Tuned hyperparameter k: {}".format(knn_cv.best_params_)) 
print("Best score: {}".format(knn_cv.best_score_))
```

2 개의 하이퍼 매개 변수가있는 기타 그리드 검색 예제

첫 번째 하이퍼 매개 변수는 c: 로지스틱 회귀 정규화 매개 변수입니다.
C가 높으면 : overfit
C가 낮 으면 : underfit
두 번째 hyperparameter는 penalty (손실 함수)입니다. l1 (Lasso) 또는 l2 (Ridge)는 선형 회귀 분석에서 배웠습니다.

```python
# 2 hyperparameter를 가진 grid search cross validation 
# 1. hyperparameter는 C:logistic regression regularization parameter입니다.
# 2. 패널티는 l1 or l2 
# Hyperparameter grid
param_grid = {'C': np.logspace(-3, 3, 7), 'penalty': ['l1', 'l2']}
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 12)
logreg = LogisticRegression()
logreg_cv = GridSearchCV(logreg,param_grid,cv=3)
logreg_cv.fit(x_train,y_train)

# 최적의 매개 변수와 최상의 점수를 인쇄하십시오
print("Tuned hyperparameters : {}".format(logreg_cv.best_params_))
print("Best Accuracy: {}".format(logreg_cv.best_score_))
```

KMEANS
KMeans Cluster 인 감독되지 않은 첫 번째 방법을 시도해 봅니다.
KMeans 클러스터 :이 알고리즘은 제공된 기능을 기반으로 K 개의 그룹 중 하나에 각 데이터 포인트를 반복적으로 할당합니다. 데이터 포인트는 피쳐 유사성을 기반으로 클러스터됩니다.
KMeans (n_clusters = 2) : n_clusters = 2는 2 클러스터 생성

```python
# As you can see there is no labels in data
data = pd.read_csv('./input/column_2C_weka.csv')
plt.scatter(data['pelvic_radius'],data['degree_spondylolisthesis'])
plt.xlabel('pelvic_radius')
plt.ylabel('degree_spondylolisthesis')
plt.show()
```

```python
# KMeans Clustering
data2 = data.loc[:,['degree_spondylolisthesis','pelvic_radius']]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data2)
labels = kmeans.predict(data2)
plt.scatter(data['pelvic_radius'],data['degree_spondylolisthesis'],c = labels)
plt.xlabel('pelvic_radius')
plt.xlabel('degree_spondylolisthesis')
plt.show()
```

```python
EVALUATING OF CLUSTERING

데이터를 두 그룹으로 묶습니다. Okey는 올바른 클러스터링입니까? 클러스터링을 평가하기 위해 교차 표를 사용합니다.

0 * 및 * 1 인 두 개의 클러스터가 있습니다.
1 등급 0은 비정상 인 138 명과 정상인 100 명을 포함합니다.
2 등급 1에는 72 개의 비정상적인 pati와 0 개의 정상 patiens가 포함됩니다. 
* 2 개의 cluster의 대부분은 비정상적인 환자입니다.
```

```python
새로운 질문은 클래스 데이터의 수를 알 수 있지만 클래스의 수가 데이터에서 알 수없는 경우는 무엇인가하는 것입니다. 이것은 KNN 또는 회귀 분석에서 하이퍼 파라미터와 비슷합니다.

관성(inertia) : 클러스터 밖으로 퍼져 나가는 것이 각 샘플로부터의 거리
낮은 관성은 더 많은 클러스터를 의미합니다.
최고의 클러스터 수는 무엇입니까? * 낮은 inertia 와 너무 많은 클러스터 교환이 없으므로 elbow를 선택할 수 있습니다.
```

```python
# inertia
inertia_list = np.empty(8)
for i in range(1,8):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data2)
    inertia_list[i] = kmeans.inertia_
plt.plot(range(0,8),inertia_list,'-o')
plt.xlabel('Number of cluster')
plt.ylabel('Inertia')
plt.show()
```

PRINCIPLE COMPONENT ANALYSIS (PCA)
기초 치수 감소 기술
첫 번째 단계는 무의식적이다 :
 데이터 샘플을 축과 정렬되도록 회전
 데이터 제곱을 평균 제로로 바꿉니다.
 정보 손실 없음
 fit () : 샘플 이동 방법 배우기
 transform () : 학습 된 변환을 적용합니다. 또한 테스트 데이터를 적용 할 수도 있습니다.
결과 PCA 피쳐는 선형 적으로 상관되지 않습니다.
주요 구성 요소 : 분산의 방향

```python
# PCA
from sklearn.decomposition import PCA
model = PCA()
model.fit(data3)
transformed = model.transform(data3)
print('Principle components: ',model.components_)
```

```python
# PCA variance
scaler = StandardScaler()
pca = PCA()
pipeline = make_pipeline(scaler,pca)
pipeline.fit(data3)

plt.bar(range(pca.n_components_), pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.show()
```

```python
두 번째 단계 : 본질적인 차원 : 차원 축소의 기본 개념 인 데이터를 근사화하는 데 필요한 기능의 수
PCA는 샘플에 여러 가지 기능이있을 때 고유 차원을 식별합니다.
본질적인 차원 = 상당한 차이가있는 PCA 피쳐의 수
고유 차원을 선택하려면 모든 것을 시도하고 최상의 정확도를 찾으십시오.
또한이 예를 통해 PCA의 직관적 인 방법을 확인하십시오. https://www.kaggle.com/kanncaa1/tutorial-pca-intuition-and-image-completion
```
