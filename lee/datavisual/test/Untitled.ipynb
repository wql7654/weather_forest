```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings # ignore warnings
warnings.filterwarnings("ignore")
```

```python
from collections import Counter
%matplotlib inline
```

```python
data = pd.read_csv('../pokemon.csv')
```

```python
#  
# color = color, label = label, linewidth =  , alpha = , grid = , linestyle =  
data.Attack.plot(kind = 'line', color = 'b',label = 'Attack',linewidth=1,alpha = 0.5,grid = True,linestyle = '-.')
data.Defense.plot(kind = 'line',color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     # legend =   .
plt.xlabel('x axis')              # label =  
plt.ylabel('y axis')
plt.title('Line Plot')            # title =  
plt.show()
```

```python
# 
# x = , y = 
data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red')
plt.xlabel('Attack')              # label = name of label
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')            # title = title of plot
```

```python
# 
# bins =   
data.Defense.plot(kind = 'hist',bins = 50,figsize = (12,12))
data.Attack.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.legend()
plt.show()
```

```python
data.corr().T.style.background_gradient(cmap='summer_r')
```

```python
#  
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax ,cmap='summer_r')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(9, 5))
sns.kdeplot(data[data['Legendary'] == True]['Attack'], ax=ax)
sns.kdeplot(data[data['Legendary'] == False]['Attack'], ax=ax)
sns.kdeplot(data[data['Legendary'] == True]['Defense'], ax=ax)
sns.kdeplot(data[data['Legendary'] == False]['Defense'], ax=ax)
plt.legend(['Legendary == 1', 'Legendary == 0','Legendaryw == 0', 'Legendaryw == 1'])
plt.show()
```

```python
# plt.figure(figsize=(14,6))
sns.set_style('whitegrid')
sns.lmplot(
    x="Attack",
    y="Defense",
    data=data,
    fit_reg=False,
    hue='Legendary',
    palette="Set1")
```

```python
y_position=1.0
f, ax = plt.subplots(2, 1, figsize=(18, 8))
data['Attack'].value_counts().plot.bar(color=['r','g','b'], ax=ax[0])
ax[0].set_title('Attack by count', y=y_position)
ax[0].set_ylabel('Count')
sns.countplot('Attack', hue='Legendary', data=data, ax=ax[1])
ax[1].set_title('Attack by count legendary', y=y_position)
plt.show()
```
