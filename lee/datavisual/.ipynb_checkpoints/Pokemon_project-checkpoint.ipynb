```python
x=data_pok['Sp. Atk']
y=data_pok['HP']
fp1=np.polyfit(x,y,1) #x,y,degree(차수)
fp2=np.polyfit(x,y,2) #x,y,degree(차수)
fp3=np.polyfit(x,y,3) #x,y,degree(차수)
```

```python
fx1=np.poly1d(fp1)
fx2=np.poly1d(fp2)
fx3=np.poly1d(fp3)
```

```python
data_pok['REG_1']=fx1(x)
data_pok['REG_2']=fx2(x)
data_pok['REG_3']=fx3(x)
```

```python
plt.plot(y,x,'r.')
plt.plot(data_pok['REG_1'],x,'b.')
plt.plot(data_pok['REG_2'],x,'c.')
plt.plot(data_pok['REG_3'],x,'g.')

plt.ylabel('Sp. Atk')
plt.xlabel('HP')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize = (12,12))

ax1 = plt.subplot2grid((8,8), (0,0), rowspan=3, colspan=3) 
ax2 = plt.subplot2grid((8,8), (4,0), rowspan=3, colspan=3)
ax3 = plt.subplot2grid((8,8), (0, 4), rowspan=3, colspan=3)
ax4 = plt.subplot2grid((8,8), (4, 4), rowspan=3, colspan=3)

fig.tight_layout()

ax1.set_title("Plot1: HP and Attack", fontsize =18)
ax2.set_title("Plot2: HP and Attack", fontsize =18)
ax3.set_title("Plot3: HP and Attack", fontsize =18)
ax4.set_title("Plot4: HP and Attack", fontsize =18)

sns.regplot(x='Sp. Attack', y='Attack', 
              data=data_pok, ax=ax1) 

sns.regplot(x='Sp. Attack', y='Attack', 
              data=data_pok, fit_reg = False, color = 'green', marker ="^", ax=ax2)

sns.regplot(x='Sp. Attack', y='Attack', 
              data=data_pok, fit_reg = True, x_bins = 6, color = 'orange', ax=ax3)

sns.regplot(x='Sp. Attack', y='Attack', 
              data=data_pok, fit_reg = False, x_bins = 12, ci = 99, color = 'red', ax=ax4)


plt.show() 
```
