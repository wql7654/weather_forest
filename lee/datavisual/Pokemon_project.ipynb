```python
x=data_pok['Sp. Atk']
y=data_pok['HP']
fp1=np.polyfit(x,y,1) #x,y,degree()
fp2=np.polyfit(x,y,2) #x,y,degree()
fp3=np.polyfit(x,y,3) #x,y,degree()
```

```python
fx1=np.poly1d(fp1)
fx2=np.poly1d(fp2)
fx3=np.poly1d(fp3)
```

```python
data_pok['REG_1']=fx1(x)
data_pok['REG_2']=fx2(x)
data_pok['REG_3']=fx3(x)
```

```python
plt.plot(y,x,'r.')
plt.plot(data_pok['REG_1'],x,'b.')
plt.plot(data_pok['REG_2'],x,'c.')
plt.plot(data_pok['REG_3'],x,'g.')

plt.ylabel('Sp. Atk')
plt.xlabel('HP')
plt.show()
```

```python
fig, axs = plt.subplots(4, 6, figsize = (10,10))  

ax1 = plt.subplot2grid((4, 6), (0, 0), rowspan=2, colspan=2) 
ax2 = plt.subplot2grid((4, 6), (0, 2), rowspan=2, colspan=2)
ax3 = plt.subplot2grid((4, 6), (0, 4), rowspan=2, colspan=2)
ax4 = plt.subplot2grid((4, 6), (2, 0), rowspan=2, colspan=2)
ax5 = plt.subplot2grid((4, 6), (2, 2), rowspan=2, colspan=2)
ax6 = plt.subplot2grid((4, 6), (2, 4), rowspan=2, colspan=2)

fig.tight_layout(pad = 0.4, w_pad = 3.0, h_pad = 4.0) 

ax1.set_title("Sp. Atk verse Attack", fontsize =16)
ax2.set_title("Sp. Atk verse Speed", fontsize =16)
ax3.set_title("Sp. Atk verse Defense", fontsize =16)
ax4.set_title("Sp. Atk verse HP", fontsize =16)
ax5.set_title("Sp. Atk verse Sp. Def", fontsize =16)
ax6.set_title("Sp. Atk verse Total", fontsize =16)

sns.regplot(x='Sp. Atk', y='Attack', 
              data=data_pok, x_bins = 12, fit_reg = True, ci = 90, color = 'red', ax=ax1) 

sns.regplot(x='Sp. Atk', y='Defense', 
              data=data_pok, x_bins = 12, fit_reg = True, ci = 90, color = 'blue', ax=ax2)

sns.regplot(x='Sp. Atk', y='Speed', 
              data=data_pok, x_bins = 12, fit_reg = True, ci = 90, color = 'green', ax=ax3)

sns.regplot(x='Sp. Atk', y='HP', 
              data=data_pok, x_bins = 12, fit_reg = True, ci = 90, color = 'orange', ax=ax4)

sns.regplot(x='Sp. Atk', y='Sp. Def', 
              data=data_pok, x_bins = 12, fit_reg = True, ci = 90, color = 'purple', ax=ax5)

sns.regplot(x='Sp. Atk', y='all_data', 
              data=data_pok, x_bins = 12, fit_reg = True, ci = 90, color = 'brown', ax=ax6)

sns.set(style="whitegrid") 
sns.despine(left=True) 
sns.set_context("paper") 

plt.show()
```
