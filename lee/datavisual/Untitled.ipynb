```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('womens.csv',header=0)
data.columns = data.columns.str.replace(' ','')
```

```python
data.info()
```

```python
data = data.drop('Unnamed:0',axis=1)
data = data.drop('Title',axis=1)
data = data.drop('ReviewText',axis=1)
```

```python
data['Age_map'] = 0
data.loc[data['Age'] < 30, 'Age_map'] = 0
data.loc[(30 <= data['Age']) & (data['Age'] < 40), 'Age_map'] = 1
data.loc[(40 <= data['Age']) & (data['Age'] < 50), 'Age_map'] = 2
data.loc[(50 <= data['Age']) & (data['Age'] < 60), 'Age_map'] = 3
data.loc[data['Age'] >= 60, 'Age_map'] = 4
data
```
